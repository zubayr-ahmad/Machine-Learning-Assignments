### Run Predictions (Testing data)

```python
# Your code Here
# In order to calculate the pridicted values of y
# Use the formula y = 0_node + 0_1 * x
# _____DO NOT USE FOR LOOP_____
# instead Convert the column of test_x into numpy a 2 dimensional matrix
# then use matrix multiplication to calculate the predicted values of y

w0 = X[0]
w1 = X[1]

test_data = pd.read_csv('testRegression.csv')
x_test = np.array(test_data['X'])
y_test = np.array(test_data['R'])

y_pred = w0 + w1 * x_test
```

### Now You Are Required To Fit Your 4 , 5 and 6 degree Models Step By Step As You Did Before

## Now Fit The 4 degree Model

```python
#Your code Here

A = np.array([[x_train.shape[0], x_train.sum(), (x_train**2).sum(), (x_train**3).sum(), (x_train**4).sum()],
            [x_train.sum(), (x_train**2).sum(), (x_train**3).sum(), (x_train**4).sum(), (x_train**5).sum()],
            [(x_train**2).sum(), (x_train**3).sum(), (x_train**4).sum(), (x_train**5).sum(), (x_train**6).sum()],
            [(x_train**3).sum(), (x_train**4).sum(), (x_train**5).sum(), (x_train**6).sum(), (x_train**7).sum()],
            [(x_train**4).sum(), (x_train**5).sum(), (x_train**6).sum(), (x_train**7).sum(), (x_train**8).sum()]])
B = np.array([y_train.sum(),np.dot(x_train,y_train).sum(), np.dot(x_train**2,y_train).sum(), np.dot(x_train**3,y_train).sum(), np.dot(x_train**4,y_train).sum()]).reshape(5,1)
print("Matrix A :")
print(A)
print("Matrix B :")
print(B)

X = np.dot(np.linalg.inv(A),B)
print("Matrix X :")
print(X)

w0 = X[0]
w1 = X[1]
w2 = X[2]
w3 = X[3]
w4 = X[4]

y_pred = w0 + w1 * x_test + w2 * (x_test**2) + w3 * (x_test**3) + w4 * (x_test**4)

mse = (1/(2*x_test.shape[0]))*(np.sum((y_test - y_pred)**2))
print("Mean Square Error",mse)

plt.title("4th degree model for train Data")
plt.plot(x_train,y_train,'o')
plt.plot(x_train, w0 + w1 * x_train + w2 * (x_train**2) + w3 * (x_train**3) + w4 * (x_train**4))
plt.legend(['Train Data','4th degree Model'])
plt.show()
plt.title("4th degree model for test Data")
plt.plot(x_test, y_test,'o')
plt.plot(x_test, y_pred)
plt.legend(['Test Data','4th degree Model'])
plt.show()


```

## Now Fit The 6 degree Model
